```python
shark_df.columns = shark_df.columns.str.strip()    #strip whitespaces
shark_df = shark_df.dropna(axis=1, how='all')      #delete empty columns
```

```python
# drop weird column names
shark_df= shark_df.drop(columns=['Unnamed: 11', 'Unnamed: 22', 'pdf', 'href formula','href','Case Number','Case Number.1','original order','Unnamed: 21'])
```

```python
# capitalize country
shark_df['Country'] = shark_df['Country'].str.capitalize()
```

```python
# delete all empty rows
shark_df.dropna(how='all', inplace=True)
```

```python
# count the number of cases with 'FATAL'
fatal_count = shark_df['Injury'].str.contains(r'Fatal', regex=True, case=False).sum()

print(f"Number of cases with 'FATAL': {fatal_count}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the activity grouping function
def activity_grouping(x):
    if isinstance(x, str):
        act = x.lower()
        if "fish" in act:
            return "Fishing"
        if "board" in act or "surf" in act:
            return "Surfing"
        if "swim" in act:
            return "Swimming"
    return "Other"

# Apply the function to create a new column for Activity Type
shark_df['Activity Type'] = shark_df['Activity'].apply(activity_grouping)

# Calculate fatal and non-fatal counts by activity type
shark_df['Is Fatal'] = shark_df['Injury'].str.contains(r'Fatal', regex=True, case=False, na=False)

activity_counts = shark_df.groupby(['Activity Type', 'Is Fatal']).size().unstack(fill_value=0).reset_index()

activity_counts.columns = ['Activity Type', 'Non-Fatal', 'Fatal']

activity_counts = activity_counts.melt(id_vars='Activity Type', var_name='Injury Type', value_name='Count')

plt.figure(figsize=(10, 6))
sns.barplot(x='Activity Type', y='Count', hue='Injury Type', data=activity_counts, palette='pastel')

plt.title("Counts of Fatal and Non-Fatal Shark Incidents by Activity", fontsize=16)
plt.xlabel("Activity Type", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.legend(title='Injury Type')

plt.show()
```

```python
# Apply the daytime_grouping function
def daytime_grouping(x):
    if isinstance(x, str):
        temp = x

        # Check for specific words first
        if "Afternoon" in temp:
            return "Afternoon"
        elif "Morning" in temp:
            return "Morning"
        elif "Night" in temp:
            return "Night"
        elif "Evening" in temp:
            return "Evening"

        # If the input contains 'h', process the time-based grouping
        if "h" in temp:
            hour = temp.split("h")
            if not hour[0].isdigit():
                return "Ungrouped"
            hour_int = int(hour[0])
            if 6 <= hour_int < 12:
                return "Morning"  # from 06 to 12 (exclusive)
            elif 12 <= hour_int < 15:
                return "Afternoon"  # from 12 to 15 (exclusive)
            elif 15 <= hour_int < 18:
                return "Day"  # from 15 to 18 (exclusive)
            elif 18 <= hour_int <= 24:
                return "Evening"  # from 18 to 24/00
            elif 0 <= hour_int < 6:
                return "Night"  # from 00 to 06 (exclusive)
            else:
                return "Ungrouped" 

# Apply the function to the 'Time' column
shark_df['Daytime_Group'] = shark_df['Time'].apply(daytime_grouping)

# Plot the results
group_counts = shark_df['Daytime_Group'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))
group_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Daytime Groups')
plt.xlabel('Daytime Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
shark_df['Year'] = shark_df['Year'].apply(lambda x: str(x).rstrip('.0'))
shark_df.dropna(subset=['Date'], inplace=True)
shark_df['Date'] = pd.to_datetime(shark_df['Date'],format='%Y%m%d', errors='coerce')
shark_df['Date'] = shark_df['Date'].dt.strftime('%d-%m-%Y')
shark_df['month'] = shark_df['Date'].apply(lambda x: str(x).split("-")[1])
shark_df['month'].value_counts().sort_index()
pivot_table = shark_df.pivot_table(index='month', aggfunc='size')
pivot_table
pivot_table.plot(kind='bar', figsize=(10, 6))
# Add a title and labels
plt.title('Seasonal correlation')
plt.ylabel('Numbers of attacks')
plt.xlabel('Months of the year')
# Show the plot
plt.show()
```
